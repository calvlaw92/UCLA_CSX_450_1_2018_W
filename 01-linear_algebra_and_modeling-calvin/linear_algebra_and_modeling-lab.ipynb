### 1. Use Vectors to Plot these functions

- $y=-4x-5$
- $y=2x+16$
- $y=x^2-3x+2$
- $4x-5y=10$

```r
xx <- seq(-20,20, by=0.2)
yy1 <- -4*xx - 5
yy2 <- 2*xx + 16
yy3 <- xx^2 -3*xx + 2
yy4 <- -(10-4*xx)/5

library(repr)

options(repr.plot.width=12, repr.plot.height=6)
plot(xx, yy1, ylim=c(-20,30), xlim=c(-20,8), xlab="X", ylab="Y", col="red")
points(xx, yy2, col="green")
points(xx, yy3, col="blue")
points(xx, yy4, col="purple")
abline(h=0); abline(v=0)
```

### 4. Calculate and draw a vector addition

Find the result of the vector addition:

$$\mathbf{g}=\mathbf{a}+\mathbf{b}+\mathbf{a}$$

```r
a <- c(1,2)
b <- c(3,0.5)
g <- a + b + a
```

```r
g
```

```r
vecplot <- function (v, origin=NULL, col="black", lwd=1) {
    if (is.null(origin)) {
        origin = c(0,0)
    }
    arrows(origin[1],origin[2],v[1]+origin[1],v[2]+origin[2], col=col, lwd=lwd)
    }
```

```r
library(repr)
```

```r
options(repr.plot.width=12, repr.plot.height=6)
plot(NA, xlim=c(-2,5), ylim=c(-2,10), xlab="X", ylab="Y")
abline(h=0); abline(v=0)
vecplot(g, col="red")
vecplot(a+a, col="blue")
vecplot(b,a+a, col="green")
```
